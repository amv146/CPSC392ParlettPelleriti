```python
import numpy as np
import pandas as pd
from plotnine import *
from sklearn.linear_model import LinearRegression # Linear Regression Model
```

```python
# no relationship but linear
x = np.random.normal(loc = 0, scale = 1, size = 1000)
y = x*0 + np.random.normal(loc = 0, scale = 1, size = 1000)

df = pd.DataFrame({"x": x,
                  "y": y})

(ggplot(df, aes(x = "x", y = "y")) + geom_point() + theme_bw())
```

```python
# fit a linear regression
lr = LinearRegression()
lr.fit(x.reshape(-1, 1),y)

pred = lr.predict(x.reshape(-1, 1))

assump = pd.DataFrame({"pred": pred, "error" : df['y']-pred})

ggplot(assump, aes(x = "pred", y = "error")) + geom_point() + theme_minimal() + geom_smooth(method = "lm", color = "red")
```
